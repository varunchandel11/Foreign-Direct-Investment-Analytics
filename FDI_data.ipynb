```python
# Calculate growth rate for each sector
sector_growth_rate = df.drop('Sector', axis=1).pct_change(axis=1) * 100

print("Sector Growth Rate:")
print(sector_growth_rate)
```

```python
import numpy as np

# Replace infinities with NaNs
sector_growth_rate.replace([np.inf, -np.inf], np.nan, inplace=True)

# Replace remaining NaNs with 0
sector_growth_rate.fillna(0, inplace=True)

print("Sector Growth Rate (After Cleaning):")
print(sector_growth_rate)
```

```python
# Calculate average growth rate for each sector
avg_sector_growth_rate = sector_growth_rate.mean(axis=1)

print("Average Growth Rate for Each Sector:")
print(avg_sector_growth_rate)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define years
years = range(2000, 2018)

# Generate hypothetical FDI data for India
fdi_data_india = np.random.randint(5000, 20000, size=len(years))

# Generate hypothetical FDI data for United States
fdi_data_us = np.random.randint(10000, 30000, size=len(years))

# Generate hypothetical FDI data for China
fdi_data_china = np.random.randint(8000, 25000, size=len(years))

# Generate hypothetical FDI data for European Union
fdi_data_eu = np.random.randint(12000, 35000, size=len(years))

# Create DataFrame
fdi_df = pd.DataFrame({
    'Year': years,
    'India': fdi_data_india,
    'United States': fdi_data_us,
    'China': fdi_data_china,
    'European Union': fdi_data_eu
})

# Set Year column as index
fdi_df.set_index('Year', inplace=True)

# Plot FDI trends
plt.figure(figsize=(12, 6))
fdi_df.plot(kind='line', marker='o')
plt.title('FDI Trends: India vs. Other Countries')
plt.xlabel('Year')
plt.ylabel('FDI (Million $)')
plt.grid(True)
plt.legend(title='Country/Region')
plt.xticks(years)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Hypothetical FDI data for India and other countries/regions
data = {
    'Year': range(2000, 2018),
    'India': [2000, 2100, 2200, 2300, 2400, 2500, 2600, 2700, 2800, 2900, 3000, 3100, 3200, 3300, 3400, 3500, 3600, 3700],
    'US': [5000, 5100, 5200, 5300, 5400, 5500, 5600, 5700, 5800, 5900, 6000, 6100, 6200, 6300, 6400, 6500, 6600, 6700],
    'China': [3000, 3100, 3200, 3300, 3400, 3500, 3600, 3700, 3800, 3900, 4000, 4100, 4200, 4300, 4400, 4500, 4600, 4700],
    'EU': [4000, 4100, 4200, 4300, 4400, 4500, 4600, 4700, 4800, 4900, 5000, 5100, 5200, 5300, 5400, 5500, 5600, 5700]
}

# Create DataFrame
fdi_df = pd.DataFrame(data)

# Set 'Year' column as index
fdi_df.set_index('Year', inplace=True)

# Display the first few rows of the DataFrame
print(fdi_df.head())
```

```python
# Calculate initial FDI values for each country/region
initial_fdi = fdi_df.iloc[0]

# Calculate relative growth rates
relative_growth_rates = (fdi_df - initial_fdi) / initial_fdi * 100

# Plot relative growth rates with rotated x-axis labels
plt.figure(figsize=(12, 6))
relative_growth_rates.plot(kind='line', marker='o')
plt.title('Relative Growth Rates of FDI: India vs. Other Countries')
plt.xlabel('Year')
plt.ylabel('Relative Growth Rate (%)')
plt.grid(True)
plt.legend(title='Country/Region')
plt.xticks(fdi_df.index, rotation=45)  # Rotate x-axis labels by 45 degrees
plt.tight_layout()
plt.show()
```

```python


# Normalize FDI data to a common base year
base_year = 2000
normalized_fdi_data = fdi_df / fdi_df.loc[base_year] * 100

# Plot comparative trends
plt.figure(figsize=(12, 6))
plt.plot(normalized_fdi_data.index, normalized_fdi_data['India'], label='India', marker='o')

# Plot FDI trends for comparison countries
#     plt.plot(normalized_fdi_data.index, country_df / country_df.loc[base_year] * 100, marker='o')

plt.title('Comparative Trends of FDI: India vs. Other Countries/Regions')
plt.xlabel('Year')
plt.ylabel('Normalized FDI')
plt.grid(True)
plt.legend()
plt.xticks(normalized_fdi_data.index, rotation=45)
plt.tight_layout()
plt.show()
```
